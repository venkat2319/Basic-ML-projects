```python
from os import listdir

neg=[]
mypath = r'neg/'
for f in listdir(mypath):
    neg.append(f)
    
pos=[]
mypath = r'pos/'
for f in listdir(mypath):
    pos.append(f)
```

```python
len(neg),len(pos)
```

```python
negative_review=[]
for filename in neg:
    my_file = open('neg/'+filename, "r")
    content = my_file.read()
    negative_review.append(content)
```

```python
positive_review=[]
for filename in pos:
    my_file = open('pos/'+filename, "r")
    content = my_file.read()
    positive_review.append(content)
```

```python
neg_label=[]
pos_label=[]
for i in range(0,1000):
    neg_label.append(0)
    pos_label.append(1)
```

```python
len(negative_review),len(neg_label),len(positive_review),len(pos_label)
```

```python
neg_review_data = pd.DataFrame()
neg_review_data['Review'] = negative_review.copy()
neg_review_data['label'] = neg_label.copy()
```

```python
neg_review_data.head()
```

```python
pos_review_data = pd.DataFrame()
pos_review_data['Review'] = positive_review.copy()
pos_review_data['label'] = pos_label.copy()
```

```python
pos_review_data.head()
```

```python
dataframes = [pos_review_data,neg_review_data]
dataset = pd.concat(dataframes)
```

```python
dataset.shape
```

```python
df = dataset.sample(frac = 1)
```

```python
df = df.reset_index()
```

```python
df = df.drop('index',axis=1)
```

```python
df
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import nltk
nltk.download('stopwords')
from tqdm import tqdm

ps=PorterStemmer()

data_lst=[]
for text in tqdm(df['Review'],desc="Cleaning Data "):
    text=text.lower()
    text=re.sub('[^A-Za-z]',' ',text)
    text = text.split()
    text=[ps.stem(word) for word in text if word not in set(stopwords.words('english'))]
    text=' '.join(text)
    data_lst.append(text)

clean_data=pd.DataFrame(data_lst)
df['review']=clean_data
```

```python
reviews=df['Review'].values
label=df['label'].values
```

## Using TextBlob

```python
import textblob
polarity_list=[]
for review,sentiment in zip(reviews,label):
    polarity=textblob.TextBlob(review).sentiment.polarity
    print(f"Polarity of \n {review} \n is : {polarity}")
    polarity_list.append(polarity)
```

```python
pred_label=[1 if score>=0 else 0 for score in polarity_list]
len(pred_label),len(label)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(accuracy_score(label,pred_label))
print(confusion_matrix(label,pred_label))
print(classification_report(label,pred_label))
```

# Using BOW

```python
new_data=pd.DataFrame()
new_data['reviews']=reviews
new_data['sentiments']=label
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv_train=cv.fit_transform(new_data['reviews'])
cv_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(cv_train,new_data['sentiments'],test_size=0.25,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Using TfIdfVector

```python
new_data=pd.DataFrame()
new_data['reviews']=reviews
new_data['sentiments']=label
new_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
tf_train=tf.fit_transform(new_data['reviews'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(tf_train,new_data['sentiments'],test_size=0.25,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## RandomForest

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
tf_y_pred=model.predict(x_test)
print(accuracy_score(y_test,tf_y_pred))
print(confusion_matrix(y_test,tf_y_pred))
print(classification_report(y_test,tf_y_pred))
```

## Using HyperParameter Tuning - ADABOOST

```python
from sklearn.model_selection import GridSearchCV
params={'n_estimators':[100,130,150],
        'max_features':['auto','sqrt'],
         'n_jobs':[-1,1]}
model=RandomForestClassifier()

grid=GridSearchCV(model,param_grid=params,cv=5)
grid.fit(x_train,y_train)
print(grid.best_score_)
grid.best_params_
```

## Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
ada_model=AdaBoostClassifier()
ada_model.fit(x_train,y_train)
tf_y_pred=ada_model.predict(x_test)
print(accuracy_score(y_test,tf_y_pred))
print(confusion_matrix(y_test,tf_y_pred))
print(classification_report(y_test,tf_y_pred))
```

## Using HyperParameter Tuning - ADABOOST

```python
from sklearn.model_selection import GridSearchCV
params={'n_estimators':[50,70,90,110],
        'learning_rate':[0.3,0.5,0.8,1.0]}
ada_model=AdaBoostClassifier()

grid=GridSearchCV(ada_model,param_grid=params,cv=5)
grid.fit(x_train,y_train)
print(grid.best_score_)
grid.best_params_
```
